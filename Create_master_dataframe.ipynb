```python
import numpy as np
import os
import pandas as pd
```

```python
# Make dataframes for each metro's CSV file
ATL = pd.read_csv(filepath_or_buffer="Data/Metro_area_data/TicketData_ATL.csv")
AUS = pd.read_csv(filepath_or_buffer="Data/Metro_area_data/TicketData_AUS.csv")
BOS = pd.read_csv(filepath_or_buffer="Data/Metro_area_data/TicketData_BOS.csv")
CHI = pd.read_csv(filepath_or_buffer="Data/Metro_area_data/TicketData_CHI.csv")
DAL = pd.read_csv(filepath_or_buffer="Data/Metro_area_data/TicketData_DAL.csv")
DEN = pd.read_csv(filepath_or_buffer="Data/Metro_area_data/TicketData_DEN.csv")
HOU = pd.read_csv(filepath_or_buffer="Data/Metro_area_data/TicketData_HOU.csv")
LAX = pd.read_csv(filepath_or_buffer="Data/Metro_area_data/TicketData_LAX.csv")
MIA = pd.read_csv(filepath_or_buffer="Data/Metro_area_data/TicketData_MIA.csv")
NAS = pd.read_csv(filepath_or_buffer="Data/Metro_area_data/TicketData_NAS.csv")
NYC = pd.read_csv(filepath_or_buffer="Data/Metro_area_data/TicketData_NYC.csv")
OAK = pd.read_csv(filepath_or_buffer="Data/Metro_area_data/TicketData_OAK.csv")
PHI = pd.read_csv(filepath_or_buffer="Data/Metro_area_data/TicketData_PHI.csv")
POR = pd.read_csv(filepath_or_buffer="Data/Metro_area_data/TicketData_POR.csv")
SEA = pd.read_csv(filepath_or_buffer="Data/Metro_area_data/TicketData_SEA.csv")
SFO = pd.read_csv(filepath_or_buffer="Data/Metro_area_data/TicketData_SFO.csv")
```

```python
# Concatenate all dataframes together
frames = [ATL, AUS, BOS, CHI, DAL, DEN, HOU, LAX, MIA, NAS, NYC, OAK, PHI, POR, SEA, SFO]
TicketData = pd.concat(frames)
TicketData.head()
```

```python
TicketData[TicketData['ticket_vendor'] == 'Live Nation'].head(50)
```

```python
# Check how many event rows we have and how many of them have actual face values
num_rows = TicketData['face_value'].count()
num_rows_with_face_values = TicketData[TicketData['face_value'].str.contains("\$")]['face_value'].count()
num_rows_with_sold_out_values = TicketData[(TicketData['sold_out'] == '1') | 
                                           (TicketData['sold_out'] == '0')]['sold_out'].count()
num_sold_out_events = TicketData[TicketData['sold_out'] == '1']['sold_out'].count()

print("Total # of events: %s" % num_rows)
print("Total # of events with valid ticket face values: %s" % num_rows_with_face_values)
print("Total # of events with valid sold out values: %s" % num_rows_with_sold_out_values)
print("Total # of events that are sold out: %s" % num_sold_out_events)
```

```python
# Save concatenated dataframe as CSV
TicketData.to_csv(path_or_buf="Data/MasterTicketData.csv", index=False)
```
